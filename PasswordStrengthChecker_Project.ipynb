# Performing TF-IDF on data

```python
def word_divide_char(inputs):
    character=[]
    for i in inputs:
        character.append(i)
    return character
```

```python
word_divide_char('kzde5577')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(tokenizer=word_divide_char)
```

```python
X= vectorizer.fit_transform(x)
```

```python
X
```

```python
X.shape #Vectorizer would yield more rows than usual dataframe
```

```python
vectorizer.get_feature_names() #Each vector has its own independent features
```

```python
first_document_vector = X[0]
first_document_vector
```

```python
first_document_vector.T.todense() #Todense()- Convert to matrix form, Toarray()- Convert to array form #T=Transform [Feature- Respective Importance]
```

```python
df= pd.DataFrame(first_document_vector.T.todense(), index= vectorizer.get_feature_names(), columns=['TF-IDF'] )
df
```

```python
df.sort_values(by=['TF-IDF'],ascending=False) #Sorted on basis of frequency
```

# Applying ML Algorithm- Logistic Regression 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
np.array(y_train).shape
```

```python
np.array(y_test).shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf=LogisticRegression(random_state=0, multi_class='multinomial') #Multinomial class for multiple subsets of data(0,1,2)
```

```python
clf.fit(X_train,y_train) #Gives the default parameters selected by Logistic Regression
```

```python
#ML Model cannot understand string data and hence it is converted to numerical form
```

```python
dt = np.array(['@123abcd'])
pred = vectorizer.transform(dt)
clf.predict(pred) #Gives strength of password=1
```

```python
y_pred = clf.predict(X_test)
y_pred
```

# Checking the accuracy/performance of the model

To check accuracy of model, following need to be considered in model-
1. Confusion Matrix
2. Accuracy Score
3. Classification Report

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
cm = confusion_matrix(y_test, y_pred)  #5218,93028,11501 are True Predictions and rest in matrix are false predictions
print(cm)
```

```python
ac= accuracy_score(y_test,y_pred)
print(ac) #Accuracy of model is 81.9%
```

```python
from sklearn.metrics import classification_report
cr = classification_report(y_test,y_pred)
print(cr)
```
